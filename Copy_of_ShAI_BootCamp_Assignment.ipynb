```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
# prompt: mean of TotalPayBenefits for each Year

mean_total_pay_benefits_by_year = df.groupby('Year')['TotalPayBenefits'].mean()
print(mean_total_pay_benefits_by_year)
```

```python
# prompt: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.

# Calculate basic statistics
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode().values[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

# Calculate range and standard deviation
salary_range = max_salary - min_salary
std_salary = df['TotalPay'].std()

# Print the results
print(f"Mean salary: {mean_salary}")
print(f"Median salary: {median_salary}")
print(f"Mode salary: {mode_salary}")
print(f"Minimum salary: {min_salary}")
print(f"Maximum salary: {max_salary}")
print(f"Salary range: {salary_range}")
print(f"Standard deviation: {std_salary}")
```

```python
# prompt: use pie charts to represent the proportion of employees in different departments. ignore from pie chart if df['JobTitle'].value_counts()  < 5

import matplotlib.pyplot as plt
# Filter departments with less than 5 employees
filtered_df = df[df['JobTitle'].isin(df['JobTitle'].value_counts()[df['JobTitle'].value_counts() > 500].index)]

# Get department counts
department_counts = filtered_df['JobTitle'].value_counts()

# Create pie chart
plt.figure(figsize=(10, 7))
plt.pie(department_counts, labels=department_counts.index, autopct="%1.1f%%", startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
# prompt: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

# Group by year and calculate mean salary
grouped_df = df.groupby('Year')['TotalPay'].mean()

# Print the average salaries for each year
print(grouped_df)

# Compare the average salaries across different years
print(f"Average salary in 2011: {grouped_df[2011]}")
print(f"Average salary in 2012: {grouped_df[2012]}")
print(f"Average salary in 2013: {grouped_df[2013]}")

# Group by job title and calculate mean salary
grouped_df = df.groupby('JobTitle')['TotalPay'].mean()

# Print the average salaries for each job title
print(grouped_df)

# # Compare the average salaries across different job titles
# print(f"Average salary for Police Officer III: {grouped_df['Police Officer III']}")
# print(f"Average salary for Custodian: {grouped_df['Custodian']}")
# print(f"Average salary for Firefighter: {grouped_df['Firefighter']}")
```

```python
# prompt:  Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

import matplotlib.pyplot as plt
# Calculate the correlation between salary and overtime pay
correlation = df['TotalPay'].corr(df['OvertimePay'])

# Print the correlation coefficient
print(f"Correlation between salary and overtime pay: {correlation}")

# Plot a scatter plot to visualize the relationship
plt.scatter(df['TotalPay'], df['OvertimePay'])
plt.xlabel('TotalPay')
plt.ylabel('OvertimePay')
plt.title('Scatter Plot of Salary vs. Overtime Pay')
plt.show()
```

```python
# prompt: drop overfitting from overtimepay and total pay

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df[['OvertimePay']], df['TotalPay'], test_size=0.33, random_state=42)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)

# Print the mean squared error
print(f"Mean squared error: {mse}")

# Calculate the R-squared score
r2_score = model.score(X_test, y_test)

# Print the R-squared score
print(f"R-squared score: {r2_score}")
```
